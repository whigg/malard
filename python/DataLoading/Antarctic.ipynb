```python
from MalardClient.MalardClient import MalardClient
from MalardClient.DataSet import DataSet
from MalardClient.BoundingBox import BoundingBox

from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import *
import os

client = MalardClient()

#determines the databases that store multiple datasets 
print(client.getParentDataSets())
```

```python
#retrieve the data sets for a given paren
for ds in client.getDataSets('mtngla'):
    print(ds)
    
dataSet = DataSet( 'mtngla', 'ADwithTDX', 'alaska' )
```

```python

box = client.boundingBox( dataSet )
gcs = client.gridCells(dataSet, box)


proj4 = client.getProjection(dataSet).proj4

print(proj4)
```

```python

years = range(2010,2020)
months= range(1,13)
columns = []#['lat','lon','x','y','elev','heading','demDiff','demDiffMad','wf_number','sampleNb','power','phase','phaseS','coh','time']

filters =  [{'column':'power','op':'gte','threshold':10000},{'column':'coh','op':'gte','threshold':0.8},{'column':'demDiff','op':'lte','threshold':100.0},{'column':'demDiff','op':'gte','threshold':-100.0},{'column':'demDiffMad','op':'lte','threshold':10.0}]

path = '/data/puma1/scratch/v2/malard/export/'

def extractDataByMonth(path,name,years,months,columns,filters, minX,maxX,minY,maxY, shapeFile=""):
    for y in years:
        for m in months:
            startTime=datetime(y,m,1,0,0)
            endTime=startTime+relativedelta(months=+1)-timedelta(seconds=1)
            resultInfo = client.executeQueryPolygon( dataSet, startTime, endTime, shapeFile,columns,filters)
            print(resultInfo.resultFileName)
            newName = path+name+'_'+str(y)+'-'+str(m)+'.nc'
            if resultInfo.status =='Success':
                os.rename(resultInfo.resultFileName,newName)
                print(newName)
            else:
                print(resultInfo.message)
                print( name + ' has no results for: ' + str(y) +'-' + str(m) )
            
```

```python
extractDataByMonth(path,'Thwaites',years,months,columns,filters,-1.64E+06,-1.26E+06,-7.59E+05,-3.08E+05)

extractDataByMonth(path,'Amery',years,months,columns,filters, 1.34E+06,1.86E+06,3.98E+05,1.03E+06)

minX= 8.40E+05
maxX= 1.67E+06
minY= -2.16E+06
maxY= -1.36E+06

extractDataByMonth(path,'Cook',years,months,columns,filters, minX,maxX,minY,maxY)

#SIPLE
minX=-6.53E+05
maxX=6.96E+04
minY=-8.71E+05
maxY=-1.60E+05

extractDataByMonth(path,'Siple',years,months,columns, filters, minX,maxX,minY,maxY)

#wingham
minX=1.01E+06
maxX=1.27E+06
minY=-1.20E+06
maxY=-1.04E+06

extractDataByMonth(path,'Wingham',years,months,columns,filters, minX,maxX,minY,maxY)
```

```python
## Check the tolerance and power thresholds in the saved files.

from os import listdir
from os.path import isfile, join

import MalardClient.MalardHelpers as mh


def checkFile(f, check_columns ):
    df = mh.getDataFrameFromNetCDF( f )
    
    if df['coh'].min() < 0.8 or df['powerScaled'].min() < 10000:
        return ("Fail", f)
    
    return ("Success", f)

files  = [ checkFile(join(path, f), check_columns ) for f in listdir(path) ]


for s,f in files:
    print( "File processed: {} Status {}".format(f,s))
```
